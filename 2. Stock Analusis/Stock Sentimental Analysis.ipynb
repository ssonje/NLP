# Stock Price Sentimental Analysis

```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#reading the dataset
df = pd.read_csv('Data.csv',encoding = "ISO-8859-1")
```

```python
#checking all data is correct or not
df.head(3)
```

```python
#shape of the dataset
df.shape
```

```python
#traing data set and testing data set
train = df[df['Date']<'20150101']
test = df[df['Date']>'20141231']
```

```python
#Removing punctuations
data = train.iloc[:,2:27]

#removing all colos, semicolons which are not required
data.replace("[^a-zA-Z]", " ", regex=True, inplace=True)
```

```python
#Remaining the column names for ease of access
list1 = [i for i in range(25)]

#converting integer value to string
new_index = [str(i) for i in list1]

#Assignning the name
data.columns = new_index

#checking the corrections
data.head(3)
```

```python
#lowering the headline
for index in new_index:
    data[index] = data[index].str.lower()
    
#checking the corrections    
data.head(3)
```

```python
#concatinating all the headlines to one
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
#printing first row
headlines[0]
```

```python
#importing libraries for bag of words
from sklearn.feature_extraction.text import CountVectorizer

#Importing libraries for training the model
from sklearn.ensemble import RandomForestClassifier
```

```python
#implementing bag of words
countvector = CountVectorizer(ngram_range=(2,2))
traindataset = countvector.fit_transform(headlines)
```

```python
#implementing random forest classifier
random_classifier = RandomForestClassifier(n_estimators=200,criterion='entropy')
random_classifier.fit(traindataset,train['Label'])
```

```python
#concatinating all the headlines to one from the test dataset
test_transform = []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in data.iloc[row,2:27]))

#creating bag of words
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
#to get the accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

#printing confusion matrix
matrix = confusion_matrix(test['Label'],predictions)
print(matrix)
```

```python
#printing the accuracy score
score = accuracy_score(test['Label'],predictions)
print(score)
```

```python
#printing report 
report = classification_report(test['Label'],predictions)
print(report)
```
