## Solution :- Word2vec

### It helps to maintain the semantic relationship between different Words.

### Here, word is represented as a vector of 32 or more dimension instead of a single number.

### Example:- 
    man(3,6)
    women(3.2,6.2)
    
    man is related to women because 3 and 3.2 are closure

### Steps :- 
    1.Tokenization of sentance
    2.Create histogram
    3.Take most frequent words.
    4.Create a matrix with all unique words.
        It also represent the occurrence of relation between the words.
